```python
#Importing Required modules
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
#reading data from csv file
dataset_a = np.genfromtxt('hw02_a.csv', delimiter=',')
# print('printing 1st column')
# print(dataset_a[:, 0])

#Assign first column and second column to x and y
x = dataset_a[:, 0]
y = dataset_a[:, 1]
```

```python
#Minimum and maximum value of X
max_value_of_x=np.max(x)
min_value_of_x=np.min(x)
#Minimum and Maximum value of X
print("\n****************** Minimum and Maximum value of X*******************")
print("Minimum Value of X: ", min_value_of_x)
print("Maximum Value of X: ", max_value_of_x)
print("\n")
```

```python
#Mean and Median value of Y
#mean value of y
mean_of_y = np.mean(y)
#median value of y
median_of_y = np.median(y)
print("\n*************************Mean and Median Value of Y*********************")
print("Mean Value of Y: ", mean_of_y)
print("Median Value of Y: ", median_of_y)
print("\n")
```

```python
#Profit Value
high_profit = dataset_a[(dataset_a[:, 1] > 300000)]
print("\n*****************Profit Value********************")
print("High Profit Value: \n", high_profit)
print("\n")
```

```python
# plotting
plt.title("dataset a")
plt.xlabel("investment")
plt.ylabel("profit")
plt.plot(x, y, color="blue", linestyle='dotted')
plt.show()
```
